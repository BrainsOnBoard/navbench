```python
import matplotlib.pyplot as plt
import pandas as pd
import utm

df = pd.read_csv("../datasets/rc_car/test1/coordinates.csv")

x = []
y = []
for lat, lon in zip(df['Latitude'], df['Longitude']):
    easting, northing, *_ = utm.from_latlon(lat, lon)
    x.append(easting)
    y.append(northing)

plt.plot(x, y)
plt.axis('equal');
```
