```python
import sys

sys.path.append("..")

import cv2
import navbench as nb
from navbench import imgproc as ip
import matplotlib.pyplot as plt

noise_levels = (0, 0.05, 0.1, 0.5)

preprocess = (ip.histeq, ip.to_float)
im_orig = cv2.imread("../datasets/panorama_grand_canyon.jpg", cv2.IMREAD_GRAYSCALE)
assert len(im_orig)
im_orig = ip.apply_functions(im_orig, preprocess)

for noise in noise_levels:
    im = ip.add_randomness(im_orig, noise)
    diffs = nb.ridf(im, im_orig)
    ax = nb.plot_ridf(diffs, im=im)
    ax.set_title("Noise = %g" % noise)
    ax.set_ylabel("Mean image difference (px)");
```
